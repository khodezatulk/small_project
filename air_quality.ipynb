Show the year wise avarage pollution

```python
data.head()
```

```python
data['Year']=data.Timestamp.dt.year
data
```

```python
year_avg=data.groupby('Year')['PM2.5'].mean()
year_avg
```

```python
year_avg.plot(kind='line',figsize=(15,5))
```

```python
year_avg.plot(kind='line',figsize=(15,5),style='*')
```

2/draw the area plot showing the avarage polution month wise

```python
data.head()
```

```python
data['Month']=data.Timestamp.dt.month
data
```

```python
month_avg=data.groupby('Month')['PM2.5'].mean()
month_avg
```

```python
month_avg.plot(kind='area',figsize=(16,5))
```

```python
month_avg.plot(kind='line',figsize=(16,5),style='+')
```

3/draw a ber graph to show the hour wise avarage pollution

```python
data
```

```python
data['Hour']=data.Timestamp.dt.hour
data
```

```python
avg_hour=data.groupby('Hour')['PM2.5'].mean()
avg_hour
```

```python
avg_hour.plot(kind='bar',figsize=(16,6),color='purple')
```
